```python
# extracting genre without id
# and converting above string of list to list
import ast
def convert(obj):
    list = []
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list 
```

```python
 # applying above function
movies['genres']=movies['genres'].apply(convert)
```

```python
movies
```

```python
# applying on keywords
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# for cast we will take only top 4 elements and we don't need character's reel name
def convert2(obj):
    list = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 4:
            list.append(i['name'])
            counter+=1
        else :
            break
    return list 

```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies
```

```python
# for crew we don't need any except director
def extract_director(obj):
    list = []
    
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list.append(i['name'])
            break
    return list 
```

```python
movies['crew']=movies['crew'].apply(extract_director)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new= movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
new
```

```python
# text preprocessing 
# lower the text>>
import nltk

from nltk.corpus import stopwords

import string
from nltk.stem import WordNetLemmatizer
 
lemmatizer = WordNetLemmatizer()
def new_body(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
    
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
            
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(lemmatizer.lemmatize(i))
    
            
    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(new_body)
```

```python
new
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv = CountVectorizer()
tfidf = TfidfVectorizer(max_features=5000)
vec = tfidf.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vec)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend_for=input("Enter movie name:\n")
```

```python
recommend(recommend_for)
```
